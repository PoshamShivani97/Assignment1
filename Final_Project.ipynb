```python
class Restaurant:
    def __init__(self):
        self.Food_Dict={}
        self.ordered_item=[]
        self.Food_Id=len(self.Food_Dict) + 1
        self.Admin_Details={}
        self.User_Details={}
    #defining functionalities for Admin
    #functionality to register admin
    def Register_Admin(self):    
        self.Email=input("Please Enter Email Id for Admin :")
        self.Pswd=input("Please Enter Password for Admin : ")
        Admin_Details={"Email_Id" :self.Email, "Password":self.Pswd}
        print("\t\t ### Admin Account Created Successfully ###\t\t")
        print("Admin Details are: ")
        for i in Admin_Details :
            print(i + "\t:\t"+Admin_Details[i])
      #function to Login admin      
    def Login_Admin(self):
        print("welcome to Amma Restaurant")
        ad_Email=input("Please Enter Mail id to log in : ")  #should match with registered email id
        ad_Pswd=input("Please Enter PassWord id to log in : ")
        if (len(ad_Email)!=0 and len(ad_Pswd)!=0):
            if ad_Email==self.Email and ad_Pswd==self.Pswd:
                print("\n!!! LOGIN SUCCESSFULL !!!")
                while True:
                    print("\nEnter the Below Options\n")
                    print("\t1. ADD FOOD ITEM \n\t2. EDIT FOOD ITEM\n\t3. VIEW FOOD ITEM\n\t4. DELETE FOOD ITEM\n\t5. GO BACK")
                    z=input()
                    if z=="1":
                        self.add_food_item()
                    elif z=="2":
                        self.edit_food_item()
                    elif z=="3":
                        self.view_food_item()
                    elif z=="4":
                        self.delete_food_item()
                    elif z=="5":
                        break
                    else:
                        print("NOT A VALID OPTION ")
            else:
                print("\n!!! INCORRECT ADMIN DETAILS !!!\n")
        else:
            print("\n!!!Error : Login Creditials Should not be Empty !!!")
    def add_food_item(self):
        name=input("Enter the food name : ")
        quantity=float(input("Enter the quantity in values only : "))
        price=float(input("Enter the price in Rs only : "))
        discount=float(input("Enter the discount in Rs only : "))
        stock=float(input("Enter the available stock in values only : "))
        food_item={"Name":name,"Quantity":quantity,"Price":price,"Discount":discount,"Stock":stock}
        self.Food_ID=len(self.Food_Dict)+1
        self.Food_Dict[self.Food_ID]=food_item
        print("\n!! FOOD ITEM ADDED SUCCESSFULLY !!\n")
        print("UPDATED FOOD ITEMS :", self.Food_Dict,"\n")
    def edit_food_item(self):
        print("\tFOOD ID OF ITEMS ID'S ARE LISTED BELOW : \n\t1.Tandori Chicken\n\t2.Vegan Burger\n\t3.Truffle Cake")
        x=int(input("Enter the Food ID you want to update : \n"))
        if x in self.Food_Dict.keys():
            print("\t1. UPDATE FOOD NAME\n\t2. UPDATE QUANTITY\n\t3. UPDATE PRICE\n\t4. UPDATE DISCOUNT\n\t5. UPDATE STOCK\n")
            y= input("Enter the Number Only : ")
            if y=="1":
                self.Food_Dict[x]["Name"]=input("Updated Food name : ")
                print("\n!! Food Name Successfully Updated !!\n")
            elif y=="2":
                self.Food_Dict[x]["Quantity"]=float(input("Updated Quantity in values only : "))
                print("\n!! Quantity Successfully Updated !!\n")
            elif y=="3":
                self.Food_Dict[x]["Price"]=float(input("Updated Price in Rs only : "))
                print("\n!! Price Successfully Updated !!\n")
            elif y=="4":
                self.Food_Dict[x]["Discount"]=float(input("Updated Discount in Rs only : "))
                print("\n!! Discount Successfully Updated !!\n")
            elif y=="5":
                self.Food_Dict[x]["Stock"]=float(input("Updated Stock in values only : "))
                print("\n!! Stock Successfully Updated !!\n")
            else:
                print("!! NOT A VALID OPTION !!\n")
        else:
                print("\n!! INCORRECT FOOD ID !!\n")
    def view_food_item(self):
        print("LIST OF FOOD ITEMS : \n")
        if len(self.Food_Dict)!=0:    
            for i in self.Food_Dict:
                print(f"Food Id : {i}")
                for j in self.Food_Dict[i]:
                    print(j, ":", self.Food_Dict[i][j])
                print()
        else:
            print("!! FOOD ITEMS ARE NOT AVAILABLE !!\n")        
    def delete_food_item(self):
        print("!! WARNING !!\nFOOD ITEM WILL DELETE PERMANENTLY\n")
        print("Enter the Food ID ")
        x=int(input())
        if x in self.Food_Dict.keys():
            del self.Food_Dict[x]
            print("\n!! FOOD ITEM DELETED SUCCESSFULLY !!\n")
            print("UPDATED FOOD ITEM LIST :\n",self.Food_Dict)
        else:
            print("!! INCORRECT FOOD ID!!\n")
            
    #defining user functionalities
    
    def Register_User(self):
        user_name=input("Enter your full name : ")
        phone_no=int(input("Enter your 10 digit phone number : "))
        email=input("Enter your email id : ")
        password=input("Enter your password : ")
        address=input("Enter your address with area PIN code ")
        self.User_Details={"User Name":user_name,"Phone No.":phone_no,"Email_ID":email,"Password":password,"Address":address}
        print("\n!! ACCOUNT CREATED SUCCESSFULLY !!\n")
        print("USER DETAILS : ")
        for i in self.User_Details:
            print(i, ":", self.User_Details[i])
    def Login_User(self):
        print("welcome to Amma Restaurant")
        user_Email=input("Please Enter Mail id to log in : ")  #should match with registered email id
        user_Pswd=input("Please Enter PassWord id to log in : ")
        if (len(user_Email)!=0 and len(user_Pswd)!=0):
            if user_Email==self.User_Details["Email_ID"] and user_Pswd==self.User_Details["Password"]:
                print("\n!!! LOGIN SUCCESSFULL !!!")
                while True:
                    print("\nEnter the Below Options\n")
                    print("\t1. PLACE NEW ORDER\n\t2. ORDER HISTORY\n\t3. UPDATE PROFILE\n\t4. GO BACK")
                    z=input()
                    if z=="1":
                        self.place_order()
                    elif z=="2":
                        self.order_history()
                    elif z=="3":
                        self.update_order_details()
                    elif z=="4":
                        break
                    else:
                        print("NOT A VALID OPTION ")
            else:
                print("\n!!! INCORRECT ADMIN DETAILS !!!\n")
        else:
            print("\n!!!Error : Login Creditials Should not be Empty !!!")

    def place_order(self):
        if len(self.Food_Dict)!=0:
            print("LIST OF AVAILABLE FOOD ITEMS :\n")
            menu=[]
            for items in self.Food_Dict:
                menu.append([self.Food_Dict[items]["Name"], self.Food_Dict[items]["Quantity"],self.Food_Dict[items]["Price"]]) 
            for i in range(len(menu)):
                print(i+1,".",menu[i])
            while True:
                print("\n\t1. CONTINUE ORDER\n\t2. GO BACK\n")
                x=input()
                if x=="1":
                    print("Enter the Food number You want to ordered separated by comma")
                    y=input().split(",")
                    for i in y:
                        z=int(i)
                        if z<=len(menu):
                            self.ordered_item.append(menu[z-1])
                        else:
                            print("\nFOOD ITEM NOT AVAILABLE  : ",z)
                    print("\nLIST OF FOOD ITEMS SELECTED : \n")
                    for j in self.ordered_item:
                        print(j)
                elif x=="2":
                    break
                else:
                    print("!! NOT A VALID OPTION !!\n")
        else:
            print("\n!! FOOD ITEMS ARE NOT AVAILABLE !!\n")      

    def order_history(self):
        if len(self.ordered_item)!=0:
            print("\nLIST OF PREVIOUS ORDERED FOOD ITEMS : \n")
            for i in self.ordered_item:
                print(i)
        else:
            print("\n!! ORDER HISTORY NOT AVAILABLE !!")
    def update_order_details(self):
        for i in self.User_Details:
            print(i, ":",self.User_Details[i])
        while True:
            print("Select Below Options to Update Your Profile Details\n")
            print("\t1. UPDATE NAME\n\t2. UPDATE PHONE NO\n\t3. UPDATE EMAIL ID\n\t4. UPDATE PASSWORD\n\t5. UPDATE ADDRESS\n\t6. GO BACK\n")
            x=input()
            if x=="1":
                self.User_Details["User Name"]=input("Enter your updated full name : ")
                print("\n!! Name Successfully Updated !!\n")
            elif x=="2":
                self.User_Details["Phone No."]=int(input("Enter your updated 10 digit phone number : "))
                print("\n!! Phone No Successfully Updated !!\n")      
            elif x=="3":
                self.User_Details["Email_ID"]=input("Enter your updated email id : ")
                print("\n!! Email ID Successfully Updated!!")
            elif x=="4":
                self.User_Details["Password"]=input("Enter your updated password : ")
                print("\n!! Password Successfully Updated !!\n")
            elif x=="5":
                self.User_Details["Address"]=input("Enter your updated address with area PIN code ")
                print("\n!! Address Updated Successfully Updated!!\n")
            elif x=="6":
                break
            else:
                print("\n!! NOT A VALID OPTION !!\n")
                    
            if x in ["1","2","3",'4',"5"]:
                for i in self.User_Details:
                    print(i, ":",self.User_Details[i])
            else:
                pass      


            
            
            
def Restaurant_main():
    obj=Restaurant()
    print("="*105)
    print("*"*105)
    print("*"+"\t\t\t\t WELCOME TO AMMA RESTAURANT\t\t\t\t\t\t"+"*")
    print("*"*105)
    print("="*105)
    while True:
        print("!!! HOME PAGE !!!")
        print("\t1.Admin\n\t2.User\n\t3.Exit")
        inp = input("Enter Value to select admin or user : ")
        if inp=="1":
            while True:
                print("\n"+"-"*40+"ADMIN PAGE"+"-"*40+"\n")
                print("\t1. ADMIN REGISTRATION\n\t2. ADMIN LOGIN\n\t3. GO BACK\n")
                y = input()
                if y == "1":
                    obj.Register_Admin()
                elif y == "2":
                    obj.Login_Admin()           
                elif y == "3":
                    break
                else:
                    print("\nNOT A VALID OPTION\n")
        elif inp == "2":
            while True:
                print("\n!!! USER PAGE !!!\n")
                print("\t1. CREATE ACCOUNT\n\t2. LOGIN\n\t3. GO BACK\n")
                y = input()
                if y == "1":
                    obj.Register_User()
                elif y == "2":
                    obj.Login_User()           
                elif y == "3":
                    break
                else:
                    print("\nNOT A VALID OPTION ")        
        elif inp == "3":
            break
        else:
            print("NOT A VALID OPTION")
if __name__=="__main__":
    Restaurant_main()
    
```
