```python
class Flower:
    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price
        self.quantity = quantity
    def __str__(self):
        return f"{self.name} (Rs.{self.price}/ea)" # ea refers to each 
class Bouquet:
    def __init__(self, name, flowers):
        self.name = name
        self.flowers = flowers
    def __str__(self):
        return f"{self.name} bouquet: {', '.join([str(f) for f in self.flowers])}"
class FlowerShop:
    def __init__(self):
        self.flowers = []
        self.bouquets = []
    def add_flower(self, name, price, quantity):
        flower = Flower(name, price, quantity)
        self.flowers.append(flower)
    def add_bouquet(self, name, flower_quantities):
        flowers = []
        for flower_name, quantity in flower_quantities.items():
            flower = next((f for f in self.flowers if f.name == flower_name), None)
            if not flower:
                print(f"Error: {flower_name} not found in inventory")
                return
            if quantity > flower.quantity:
                print(f"Error: not enough {flower_name} in inventory")
                return
            flower.quantity -= quantity
            flowers.append(f"{quantity}x {flower.name}")
        bouquet = Bouquet(name, flowers)
        self.bouquets.append(bouquet)
    def list_inventory(self):
        print("Flower inventory:")
        for flower in self.flowers:
            print(f"{flower.name} (Rs.{flower.price}/ea) - {flower.quantity}")
    def list_bouquets(self):
        print("Available bouquets:")
        for bouquet in self.bouquets:
            print(bouquet)
    def order_flowers(self):
        for flower in self.flowers:
            if flower.quantity < 10:
                print(f"Ordering more {flower.name}...")
                flower.quantity += 50
        print("Order Complete")
# Create a flower shop object
flower_shop=FlowerShop() # Add some flowers to the inventory
flower_shop.add_flower("Rose", 2.50, 100)
flower_shop.add_flower("Lily", 1.50, 50)
flower_shop.add_flower("Tulip", 1.00, 75) # Add some bouquets to the shop
flower_shop.add_bouquet("Valentine's Day", {"Rose": 10, "Lily": 5})
flower_shop.add_bouquet("Spring Fling", {"Rose": 5, "Tulip": 10}) # List the flower inventory and available bouquets
flower_shop.list_inventory()
flower_shop.list_bouquets() # Order more flowers if needed
flower_shop.order_flowers()
flower_shop.list_inventory() # Sell a bouquet
flower_shop.add_bouquet("Mother's Day", {"Rose": 10, "Tulip": 15})
flower_shop.list_inventory()
flower_shop.list_bouquets()
```
